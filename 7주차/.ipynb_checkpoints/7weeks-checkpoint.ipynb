```python
import requests
```

```python
url = 'https://news.naver.com/main/read.naver?mode=LSD&mid=shm&sid1=100&oid=020&aid=0003403986'
```

```python
# requests.get(url) -> 안티크롤링으로 데이터수집을 막아놓음
```

```python
status = requests.get(url, headers = {'User-Agent' : 'Mozilla/5.0'}) # 우회
```

```python
status.text
```

```python
from bs4 import BeautifulSoup
```

```python
bs = BeautifulSoup(status.text, 'lxml')
```

```python
bs.select_one("#articleTitle")
```

```python
bs.select_one("#articleTitle").string
```

```python
bs.select("div#articleBodyContents._article_body_contents")
```

```python
li = bs.select("div#articleBodyContents._article_body_contents")
```

```python
for i in li:
    print(i.text)
```

```python
with open('naver_news.txt', 'w') as f:
    for i in li:
        data = i.text
        f.write(data)
```
